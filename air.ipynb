```python
# create a list with the names of numerical columns data
numerics = ['accommodates', 'bedrooms', 'minimum_nights', 'maximum_nights','review_scores_rating',
            'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin',
            'review_scores_communication', 'review_scores_location', 'review_scores_value']
#using the above columns list and price create a new dataframe
numeric_df = final_df[numerics + ['price']]

#calculate correlation matrix
correlation_matrix = numeric_df.corr()

#sorting the matrix by price in descending order
price_correlation = correlation_matrix['price'].sort_values(ascending = False)

#print the attributes that have highest correlation with price
print(price_correlation)
```

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Univariate analysis - Distribution of Listing Prices
plt.figure(figsize=(10, 6))
sns.histplot(final_df['price'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Listing Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.xlim(0, final_df['price'].quantile(0.95))  # Limiting x-axis to 95th percentile for better visibility
plt.show()
```

This histogram shows the distribution of listing prices, mostly concentrated at lower prices, as indicated by the taller bars on the left. The plot, limited to the 95th percentile, excludes high-price outliers for clarity. The overlaying sky blue curve represents the price distribution's smooth trend.

```python
# Bivariate analysis - Price vs. Property Type
plt.figure(figsize=(12, 8))
sns.boxplot(data=final_df, x='price', y='property_type')
plt.title('Price Distribution by Property Type')
plt.xlabel('Price')
plt.ylabel('Property Type')
plt.xlim(0, final_df['price'].quantile(0.95))  # Limiting x-axis to 95th percentile for better visibility
plt.show()
```

This boxplot illustrates how listing prices vary across different property types. Each box represents the interquartile range of prices for a given property type, with the median price indicated by a line within the box. Whiskers extend to show the range of typical prices, excluding outliers. The plot is restricted to the 95th percentile of prices to focus on common price ranges and improve readability by excluding extreme values.

```python
# Bivariate analysis - Price vs. Room Type
plt.figure(figsize=(10, 6))
sns.boxplot(data=final_df, x='price', y='room_type')
plt.title('Price Distribution by Room Type')
plt.xlabel('Price')
plt.ylabel('Room Type')
plt.xlim(0, final_df['price'].quantile(0.95))  # Limiting x-axis to 95th percentile for better visibility
plt.show()
```

This boxplot compares listing prices across various room types, with boxes representing the middle 50% of prices for each room type and the median price marked inside. The plot is confined to the 95th percentile of prices to exclude extreme outliers and enhance clarity.

```python
# count the hosts joined each month
hosts_per_month = final_df['date'].value_counts().sort_index()

#plot the trend of hosts joining 
plt.figure(figsize = (10,6))
hosts_per_month.plot(kind = 'line', marker = '.', color = '#051399')
plt.title('Number of Hosts Joined Airbnb Overtime')
plt.xlabel('Year-Month')
plt.ylabel('Number of Hosts Joined')
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

This line graph tracks the number of Airbnb hosts joining over time, plotted monthly. The x-axis represents the year and month, while the y-axis shows the count of new hosts. The line, marked with dots and colored in deep blue, illustrates the trend of host sign-ups, with grid lines enhancing readability. The plot indicates changes in Airbnb's popularity or host recruitment efforts over the observed period.

From the above graph, we can observe that there is an huge increase in the hosts who joined Airbnb during the years 2017 and 2018. And suddenly there was a downward trend after 2019.

```python
# Plot a graph to find the monthly trend of bookings for airbnb

# Convert the 'date' column to datetime format
final_df['date'] = pd.to_datetime(final_df['date'])

# Extract the month from the 'date' column
final_df['month'] = final_df['date'].dt.month

# Group by the 'month' column, count the number of bookings in each month, and sort by month
monthly_bookings = final_df.groupby('month').size()

# Plotting
plt.figure(figsize=(15, 8))
monthly_bookings.plot(kind='line', linewidth=2, marker='o', color='#e69305')
plt.title('Monthly Trend of Airbnb Bookings')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

This line graph shows the monthly trend of Airbnb bookings, with each point representing the total number of bookings for a given month. The x-axis labels months from January to December, while the y-axis counts the bookings. The orange line, highlighted with markers at each monthly point, illustrates seasonal variations or trends in booking volume throughout the year.

 In the months of September and October there are a huge number of rooms booked with airbnb.

# Task 4: Feature Engineering

```python
# Feature Engineering

# 1. Average Rating
rating_columns = ['review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin',
                  'review_scores_communication', 'review_scores_location', 'review_scores_value']
final_df['average_rating'] = final_df[rating_columns].mean(axis=1)


final_df['estimated_occupancy'] = final_df['review_scores_rating'].fillna(0) / 100  

```

```python

# To avoid division by zero, we'll replace 0 bedrooms with 1 (assuming at least 1 bedroom or studio)
final_df['bedrooms'] = final_df['bedrooms'].replace(0, 1)
final_df['price_per_bedroom'] = final_df['price'] / final_df['bedrooms']

# Overview of the new features
new_features_overview = final_df[['average_rating', 'estimated_occupancy', 'price_per_bedroom']].describe()

new_features_overview
```

```python
# Data Visualization for New Features

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Visualization - Distribution of Average Rating
plt.figure(figsize=(10, 6))
sns.histplot(final_df['average_rating'], bins=30, kde=True, color='purple')
plt.title('Distribution of Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.show()
```

This histogram illustrates the distribution of average ratings across Airbnb listings, with the x-axis representing the average rating score and the y-axis showing how many listings fall into each rating category. The data is divided into 30 bins, and a kernel density estimate (KDE) overlay, in purple, provides a smooth curve that highlights the overall trend in the distribution. The plot aims to show how ratings are spread out, indicating whether most listings are highly rated, or if there's a wide variation in guest satisfaction.

```python
# Visualization - Estimated Occupancy Rate
plt.figure(figsize=(10, 6))
sns.histplot(final_df['estimated_occupancy'], bins=30, kde=True, color='orange')
plt.title('Estimated Occupancy Rate')
plt.xlabel('Occupancy Rate')
plt.ylabel('Frequency')
plt.show()
```

This histogram shows the estimated occupancy rates for Airbnb listings, with the x-axis indicating the occupancy rate and the y-axis showing the number of listings at each occupancy level. The plot, using 30 bins and an orange color scheme, includes a kernel density estimate (KDE) to outline the general trend of occupancy rates. The purpose is to visualize how often listings are occupied, revealing any common occupancy patterns or the prevalence of highly frequented listings.

```python
# Visualization - Price per Bedroom
plt.figure(figsize=(10, 6))
sns.histplot(final_df['price_per_bedroom'], bins=50, kde=True, color='green')
plt.title('Distribution of Price per Bedroom')
plt.xlabel('Price per Bedroom')
plt.ylabel('Frequency')
plt.xlim(0, final_df['price_per_bedroom'].quantile(0.95))  # Limiting x-axis to 95th percentile for better visibility
plt.show()
```

This histogram depicts the distribution of Airbnb listing prices per bedroom, with the green bars showing the frequency of listings at various price points. The x-axis represents the price per bedroom, while the y-axis indicates how many listings fall into each price category. A kernel density estimate (KDE) is overlaid to provide a smooth depiction of the distribution trend. The x-axis is limited to the 95th percentile of the data to focus on the most common price ranges and exclude extreme outliers, enhancing the plot's readability and relevance.
